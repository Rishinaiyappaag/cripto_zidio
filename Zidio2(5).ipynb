üìà Candlestick Chart (Interactive)

```python
import pandas as pd
import plotly.graph_objects as go

def interactive_candlestick(file_name):
    df = pd.read_csv(file_name)
    df['Date'] = pd.to_datetime(df['timestamp'])
    df.sort_values('Date', inplace=True)

    fig = go.Figure(data=[go.Candlestick(
        x=df['Date'],
        open=df['open'],
        high=df['high'],
        low=df['low'],
        close=df['close'],
        increasing_line_color='green',
        decreasing_line_color='red'
    )])

    fig.update_layout(title=f"{file_name.replace('-USD.csv','')} Candlestick Chart",
                      xaxis_title="Date",
                      yaxis_title="Price (USD)",
                      xaxis_rangeslider_visible=True)

    fig.show()
```

üß≠ Heatmap (Correlation between Features)


```python
import seaborn as sns
import matplotlib.pyplot as plt

def heatmap_correlation(file_name):
    df = pd.read_csv(file_name)
    df['Date'] = pd.to_datetime(df['timestamp'])
    df = df[['open', 'high', 'low', 'close', 'volume']]

    df = df.dropna()
    corr = df.corr()

    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title(f"{file_name.replace('-USD.csv','')} Feature Correlation Heatmap")
    plt.show()
```

üîÅ Use for Multiple Coins

```python
crypto_files = ['BNB-USD.csv', 'FET-USD.csv','STMX-USD.csv','SSX-USD.csv','CCXX-USD.csv','ETN-USD.csv','LPT-USD.csv','VUSDC-USD.csv']

for file in crypto_files:
    interactive_candlestick(file)
    heatmap_correlation(file)
```
